```python
players = purchase_data_df[['SN','Age','Gender']].drop_duplicates()
gender = players.Gender.value_counts()
total = players.SN.count()

df = pd.DataFrame({'Total Count': gender,'Percentage of Players': gender/total * 100})
df['Percentage of Players'] = df['Percentage of Players'].map('{:.2f}%'.format)
df
```

```python
#Age Demographics
#Establish bins for ages
#binning day3
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100 ]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
age_grouped_df = purchase_data_df.loc[:,["SN", "Age"]].drop_duplicates()
#Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_data_df["Age Ranges"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names)
purchase_data_df
#do groupby
age_grouped = purchase_data_df.groupby("Age Ranges")
other_members = age_grouped["SN"].nunique()

#age_demographics_total = total_players_df['Age ranges'].value_counts()
age_demographics_percentage = other_members/totalplayers * 100
#Calculate the numbers and percentages by age group

age_demographics_df = pd.DataFrame({"Total Count": other_members, "Percentage of Players": age_demographics_percentage})


#Create a summary data frame to hold the results
age_demographics_df.index.name = None

age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:.2f}%".format)

age_demographics_df

#Optional: round the percentage column to two decimal points


#Display Age Demographics Table
```
